```python
from PIL import Image
import os
```

```python
base = "data/2/2.5/test_images"
frames = []
image_paths = os.listdir(base)
if '.DS_Store' in image_paths:
    image_paths.remove('.DS_Store')
image_paths = sorted(image_paths, key=lambda path: int(path.split('.')[0]))
for image_path in image_paths:
    if image_path == '.DS_Store':
        continue
    frames.append(Image.open(f"{base}/{image_path}"))
```

```python
out_name = "data/2/2.5/test.gif"
frame_duration = 75
frames[0].save(
    out_name,
    save_all=True,
    append_images=frames[1:],
    duration=frame_duration,  # time per frame
    loop=0                    # 0 = loop forever
)
```
